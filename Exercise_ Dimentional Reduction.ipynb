```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#load digits dadasets
digits=datasets.load_digits()
X= digits.data
```

```python
#standredize the features to have mean=0 and variance=1
scaler=StandardScaler()
X_scaled= scaler.fit_transform(X)

```

```python
plt.figure(figsize=[1,1])
plt.imshow(np.array(X[5]).reshape(8,8))
plt.show()
```

```python
#Apply PCA and fit the features selected
pca= PCA(n_components=2)
X_pca= pca.fit_transform(X_scaled)
```

```python
#visualize the data
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=digits.target, alpha=0.5, cmap=plt.cm.get_cmap('Pastel2',10))
plt.colorbar()
plt.xlabel('First Principal Component')
plt.ylabel('Second Principle Component')
plt.title('PCA on Digits Dataset')
plt.show()
```

```python
print('Explaine varience:',pca.explained_variance_ratio_)
```

```python
#download bosten housing dataset
#apply pca 2
#scatter plot it


# Load the Boston Housing dataset from the UCI Machine Learning Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']
data = pd.read_csv(url, delim_whitespace=True, names=names)
```

```python
#load digits dadasets
digits=datasets.load_digits()
data= digits.data

#standredize the features to have mean=0 and variance=1
scaler=StandardScaler()
data_scaled= scaler.fit_transform(data)

plt.figure(figsize=[1,1])
plt.imshow(np.array(data[5]).reshape(8,8))
plt.show()

#Apply PCA and fit the features selected
pca= PCA(n_components=2)
data_pca= pca.fit_transform(data_scaled)

#visualize the data
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=data.PRICE, alpha=0.5, cmap=plt.cm.get_cmap('summer',10))
plt.colorbar()
plt.xlabel('First Principal Component')
plt.ylabel('Second Principle Component')
plt.title('Boosten House Dataset')
plt.show()
```
